```python
df['Class'].value_counts()
```

```python
def classificao(linha):
    if linha == 0:
        return "Normal"
    else:
        return "Fraude"
    

df_copia = df.copy() 

df_copia['Class'] = df_copia['Class'].apply(classificao)


plt.Figure(figsize=(10,5))

ax = sns.countplot(data=df_copia, x='Class')
plt.title("Distribuicao da variavel alvo")
plt.xlabel("Fraude")



# Adding counts on the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')

plt.grid(False)
plt.show()
plt.show()
```
