> **Exploração base de dados**
---

```python
def classificao(linha):
    if linha == 0:
        return "Normal"
    else:
        return "Fraude"
    

df_copia = df.copy() 

df_copia['Class'] = df_copia['Class'].apply(classificao)


plt.Figure(figsize=(10,5))

ax = sns.countplot(data=df_copia, x='Class')
plt.title("Distribuicao da variavel alvo")
plt.xlabel("Fraude")



# Adding counts on the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')

plt.grid(False)
plt.show()
plt.show()
```

```python
# Separando as variáveis independentes X(features) e Y dependente (target)
# x contém as variáveis explicativas (features)
```

```python
x = df.drop(columns='Class', axis=1)
y = df['Class']
```

> **OBS:Variavel target será unica que passarar por transformação porque ela cartegorica**

```python
from sklearn.preprocessing import LabelEncoder

label_ecoder = LabelEncoder()

y = label_ecoder.fit_transform(y)
```

```python
y
```

```python
x.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(x, y, stratify=y, random_state=5)
```

```python
print(X_treino.shape)
print(y_treino.shape)
```

> **Metricas**
---

```python
from  sklearn.metrics import classification_report, precision_recall_curve, average_precision_score
```

> **MODELO PADRAO USADO COMO BASE ARVORE DE DECISAO**

```python
from sklearn.tree import DecisionTreeClassifier

arvore = DecisionTreeClassifier(random_state=5)

arvore.fit(X_treino, y_treino)

```

>**OLHAR METRICAS DE AVALIACOES**

```python
# Fazendo previsões no conjunto de teste
y_pred = arvore.predict(X_teste)

# Relatório de métricas detalhadas
print(classification_report(y_teste, y_pred, digits=4))

```

**Classe 1 (fraudes)**

- Precision: 0.8037 → quando o modelo disse que era fraude, acertou ~80% das vezes.

- Recall: 0.6992 → ele conseguiu capturar ~70% das fraudes 

- F1: 0.7478 → bom equilíbrio entre precisão e recall.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

> **Intepretabilidade do Modelo - Quantas descisoes**
---

```python
plt.figure(figsize=(15,6))
plot_tree(arvore,filled=True, class_names=['nao', 'sim'], fontsize=1, feature_names=X_treino.columns);
```

```python
print("Acurácia treino:", arvore.score(X_treino, y_treino))
print("Acurácia teste :", arvore.score(X_teste, y_teste))


# Modelo não está generilizando e ele copio padrão overfitting
```

```python
arvore = DecisionTreeClassifier(max_depth=3, random_state=5)

arvore.fit(X_treino, y_treino)
```

```python
plt.figure(figsize=(15,6))

plot_tree(arvore,filled=True, class_names=['nao', 'sim'], fontsize=7, feature_names=X_treino.columns);
```

```python
y_pred = arvore.predict(X_teste)

# Relatório de métricas detalhadas
print(classification_report(y_teste, y_pred, digits=4))
```

**Classe 1 (fraudes)**

- Precision:  0.88 → quando o modelo disse que era fraude, acertou ~80% das vezes.

- Recall: 0.62 → ele conseguiu capturar 62% das fraudes 

- F1: 0.73 → bom equilíbrio entre precisão e recall.


```python
print("Acurácia treino:", arvore.score(X_treino, y_treino))
print("Acurácia teste :", arvore.score(X_teste, y_teste))
```
